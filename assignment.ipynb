```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
```

```python
a,u = 2,4
print(a,u)
```

```python
t = np.random.randint(100, size =100)
t
```

```python
s = u*t + 0.5*a*t*t
s
```

```python
matplotlib.rcParams["figure.figsize"] = (20,10)
plt.scatter(t, s)
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Distance-Time Graph')
```

This is not a linear model so a normal y = b + a*x will not work. We need y = b + a1*x + a2*x^2 according to distance time equation.

```python
X = pd.DataFrame(t, columns=['t'])
X['t2'] = X['t']*X['t']
X
```

```python
y = s
y
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print(f'The value of a is {model.coef_[1]*2}\nThe value of u is {model.coef_[0]}')
```

# Lasso Regression

In Lasso Regression it pennalises dimensions so even if we give the equation of regression as y = b + a1*x + a2* x^2 +a3*x^3 + a4*/x and many more then also it should give the correct answer.

Lets check

```python
X['t3'] = X['t2']*X['t']
X.head()
```

```python
X['1/t'] = 1 / X['t']
X.head()
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X,y)
lasso.score(X,y)
```

```python
lasso.coef_
```

```python
lasso.intercept_
```

```python
print(f'The value of a is {lasso.coef_[1]*2}\nThe value of u is {lasso.coef_[0]}')
```

The value here is different because it tried to fit a different polynomial whose intercept is not zero so the value the a and u comes different, but it pennalised the t^2 term and the 1/t term so their weights are very small.

# Lets plot the equation of both linear regression and lasso regression for comparision

```python
LR_value = model.intercept_ + X['t']*model.coef_[0] + X['t2']*model.coef_[1]
LR_value
```

```python
Lasso_value = lasso.intercept_ + X['t']*lasso.coef_[0] + X['t2']*lasso.coef_[1] + X['t3']*lasso.coef_[2] + X['1/t']*lasso.coef_[3]
Lasso_value
```

```python
plt.scatter(t, LR_value)
plt.scatter(t, Lasso_value)
plt.scatter(t,s)
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Distance-Time Graph')
plt.legend(["Linear Regression Value", "Lasso Regression Values", "Original Values"])
```

As it can be seen from the above plot all the values of linear regression, Lasso regression, and original values overlapes each other.
