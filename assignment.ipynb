```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
```

```python
a,u = 2,4
print(a,u)
```

```python
t = np.random.randint(100, size =100)
t
```

```python
s = u*t + 0.5*a*t*t
s
```

```python
matplotlib.rcParams["figure.figsize"] = (20,10)
plt.scatter(t, s)
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Distance-Time Graph')
```

This is not a linear model so a normal y = b + a*x will not work. We need y = b + a1*x + a2*x^2 according to distance time equation.

```python
X = pd.DataFrame(t, columns=['t'])
X['t2'] = X['t']*X['t']
X
```

```python
y = s
y
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print(f'The value of a is {model.coef_[1]*2}\nThe value of u is {model.coef_[0]}')
```
